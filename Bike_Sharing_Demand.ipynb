```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, GridSearchCV, TimeSeriesSplit
from sklearn.base import BaseEstimator
```

Les fonctions suivantes permettent de mesurer l'erreur pour les différents modèles proposés par la suite. Elles mesurent soit la racine de la moyenne des carrés des erreurs (notée RMSE), soit la racine de la moyenne du carré de l'écart des logarithmes népériens entre les prédictions et les valeurs observées (notée RMSLE). Ces mesures sont détaillées plus bas.

```python
def rmse(y_pred, y_actual):
    """
    fonction de perte RMSE
    """
    diff = y_pred - y_actual
    mean_error = np.square(diff).mean()

    return np.sqrt(mean_error)
```

```python
def rmse_scorer(pred, X, y):
    """
    fonction de score RMLE utilisee pour la validation croisee
    """
    diff = pred.predict(X)[:, 2] - y[:, 2]
    mean_error = np.mean(np.square(diff))

    return np.sqrt(mean_error)
```

```python
def rmsle(y_pred, y_actual):
    """
    fonction de perte RMSLE
    """
    diff = np.log(y_pred + 1) - np.log(y_actual + 1)
    mean_error = np.square(diff).mean()

    return np.sqrt(mean_error)
```

```python
def rmsle_scorer(reg, X, y):
    """
    fonction de perte RMSLE utilisee pour la validation croisee
    """
    pred = reg.predict(X)
    if (pred + 1 < 0).any():
        print('pred < 0')
    diff = np.log(pred[:, 2] + 1) - np.log(y[:, 2] + 1)
    if (diff < 0).any():
        print('diff < 0')
    mean_error = np.square(diff).mean()

    return np.sqrt(mean_error)
```

```python
def rmsle_ind_gs_scorer(reg, X, y):
    """
    fonction de score RMSLE utilisee pour le GridSearchCV
    """
    pred = reg.predict(X)
    pred[pred < 0] = 0
    diff = np.log(pred + 1) - np.log(y + 1)
    mean_error = np.square(diff).mean()

    return - np.sqrt(mean_error)
```

```python
def squared_log_error(y_pred, y_actual):
    """
    fonction d'erreur utilises pour les graphs
    """
    diff = np.log(y_pred + 1) - np.log(y_actual + 1)
    return np.square(diff)
```

La fonction suivante permet de construire la matrice des variables explicatives pour le modele basé sur les medianes horaires et celui du Gradient Boosting. Il extrait de nombreuses variables de la date des observations (jour, année, heure, ...) et il centre et normalise les données quantitatives.

```python
def feature_engineering(draft_data):
    """
    feature_engineering permet de construire la matrice des
    variables explicatives pour le modele base sur les medianes
    et celui du Gradient Boosting
    """

    quant_cols = ['temp', 'atemp', 'humidity', 'windspeed']
    all_data = draft_data.copy()

    # Extraction des informations contenus dans la date
    # RQ : les minutes et les secondes ne sont pas extraites
    # car la précision des relevés du dataset s'arrête à l'heure 
    all_data['day'] = draft_data.index.day
    # all_data['month'] = draft_data.index.month
    all_data['year'] = draft_data.index.year
    all_data['hour'] = draft_data.index.hour
    all_data['dow'] = draft_data.index.dayofweek
    # all_data['woy'] = draft_data.index.weekofyear
    
    quant_data = draft_data[quant_cols]

    # Centrage et standardisation des données quantitatives
    all_data[quant_cols] = (quant_data - quant_data.mean(axis=0)) / quant_data.std(axis=0)

    return all_data
```

La fonction scoring permet de mesurer les erreurs de chacuns des modèles et de les représenter au cours du temps (sur toute la période de test et en fonction des heures de la journée). Ces erreurs (RMSE et RMSLE) sont mesurées à partir de validations croisees qui tiennent compte de la dimension temporelle des donnees. Les validation croisee se font alors sur 24 periodes d'un mois. Les graphique représentent le carré des erreurs des prédictions sur les 6 derniers mois à partir d'un modèle entrainé sur les 18 premiers mois.

```python
def scoring(regressor, df_features, df_y):
    """
    scoring affiche les RMSLE et RMSE
    ainsi que les graphiques représentant
    les erreurs pour les differentes predictions
    (sur la periode et par heure)
    """
    
    X, y = df_features.values, df_y.values
    # les graphiques 
    n_split = int(X.shape[0] * 0.75)
    X_train, X_test = X[:n_split], X[n_split:]
    y_train, y_test = y[:n_split], y[n_split:]
    regressor.fit(X_train, y_train)
    pred = regressor.predict(X_test)
    error = squared_log_error(pred, y_test)

    if df_features.ndim < 2:
    # pour le modele HourlyMedian les dates sont
    # dans la matrice des variables explicatives
        x_time = df_features[n_split:]
    else:
    # pour les autres elles corespondent à l'indice du df
        x_time = df_features.index[n_split:]

    gb_error = pd.DataFrame(np.hstack([pred,
                                       y_test,
                                       error]),
                            index=x_time,
                            columns=['reg_pred', 'cas_pred', 'pred',
                                     'y_reg', 'y_cas', 'y',
                                     'reg_log_err', 'cas_lor_err', 'log_error'])
    
    gb_error['hour'] = gb_error.index.hour
    # represente l'evolution des erreurs sur la periode de test
    gb_error.groupby(pd.TimeGrouper('d')).sum().plot(y=['log_error', 'reg_log_err', 'cas_lor_err'])
    # represente la moyenne des erreurs par heure de la journee
    gb_error.groupby(by='hour').mean().plot(y=['log_error', 'reg_log_err', 'cas_lor_err'])

    # les erreurs (RMSE et RMSLE) sont calculees a partir de validations croisees
    # realisees sur 24 periodes d'un mois
    rmsle_cv = cross_val_score(regressor, X, y, cv=TimeSeriesSplit(24))
    print('CV RMSLE : {0:.4f} +/-{1:.4f}'.format(rmsle_cv.mean(), rmsle_cv.std()))
    rmse_cv = cross_val_score(regressor, X, y, scoring=rmse_scorer,
                                                    cv=TimeSeriesSplit(24))
    print('CV RMSE : {0:.2f} +/-{1:.4f}'.format(rmse_cv.mean(), rmse_cv.std()))
    print('RMSLE pour le test (graph) : {0:.4f}'.format(np.sqrt(error[:, 2].mean())))
```

#### Point de comparaison : la mediane horaire

```python
class HourlyMedian(BaseEstimator):
    """
    estimateur de basique utilise comme point de comparaison :
    le predicteur associe a une heure la mediane du nombre de
    locations de velos calculee sur l'echantillon d'entrainement
    """

    def fit(self, X, y):
        """
        calcul des medianes des variables 'registered', 'casual' et 'count'
        pour chaque heure de la journee
        """
        hourly_count = pd.DataFrame(y,
                            index=X,
                            columns=['registered', 'casual', 'count'])
        hourly_count['hour'] = hourly_count.index.hour
        self.hour_median = hourly_count.groupby('hour').median()

    def predict(self, X):
        """
        prediction des variables 'registered', 'casual' et 'count'
        a partir des medianes calculees a chaque heure
        sur chacune de ces variables lors de l'entrainement
        """
        pred = np.array(list(map(lambda h: self.hour_median.ix[h], pd.DatetimeIndex(X).hour)))
        
        return pred

    def score(self, X, y):
        """
        RMSLE calculee uniquement sur la variable 'count'
        """
        diff = np.log(self.predict(X)[:, 2] + 1) - np.log(y[:, 2] + 1)
        mean_error = np.mean(np.square(diff), axis=0)

        return np.sqrt(mean_error)
```

J'utilise dans un premier temps les médianes du nombre de locations de vélos calculées sur l'échantillon d'entrainement pour chaque heure  pour prédire sur un échantillon de test. Cet estimateur relativement simple permet de calculer l'erreur  (mesurée par le RMSLE) qui servira de point de repère pour les prédictions suivantes. La RMSLE avec une validation croisée est d'environ $0.74~ (+/-0.1)$ et la CV RMSE est proche de $128~ (+/-47)$.

```python
y_cols = ['registered', 'casual', 'count']
raw_data = feature_engineering(data)
raw_data.sort_index(inplace=True)
scoring(HourlyMedian(), raw_data.index, data[y_cols])
```

#### Ridge

```python
def linear_feature_engineering(draft_features):
    """
    linear_feature_engineering
    - ajoute l'annee et l'heure pour chaque ligne
    - discretise les variables 'weather', 'season', 'hour' et 'year'
    - redimensionne les variables quantitatives
    """
    
    quant_cols = ['temp', 'humidity', 'windspeed'] 
    
    # Extraction des informations contenus dans la date
    # RQ : les minutes et les secondes ne sont pas extraites
    # car la précision des releves du dataset s'arrête à l'heure
    # le jour et le mois n'ameliore pas la prediction
    draft_features['year'] = draft_features.index.year
    draft_features['hour'] = draft_features.index.hour

    # discretisation des variables qualitatives
    feat_wd = pd.get_dummies(draft_features, columns=['weather',
                                                      'season',
                                                      'hour',
                                                      'year']) 

    # 'atemp' n'est pas integre aux donnees car cette donnee
    # est redondante (risque de compromettre l'inversibilite 
    # de la matrice dans un modèle lineaire)
    feat_df = feat_wd.drop('atemp', axis=1)
    
    quant_data = draft_features[quant_cols]
    
    # Les donnees sont redimensionnees a partir de leur valeurs minimales et maximales
    feat_df[quant_cols] = (quant_data - quant_data.min(axis=0)) \
    / (quant_data.max(axis=0) - quant_data.min(axis=0))
    
    return feat_df
```

J'utilise le Ridge pour prédire le logarithme de la demande de location de vélos car ce régresseur est relativement simple à comprendre, à interprêter et à mettre en oeuvre. Il est par ailleurs relativement efficace sur les données après qu'elle aient été linéarisées (et discrétisées lorsque c'est nécéssaire) puisqu'il permet de réduire la CV RMSLE aux alentours de $0.60$. La régularisation l2 permet d'éviter l'explosion des coefficients en réduisant significativement l'impact des valeurs extrêmes. En effet, la regression linéaire est beaucoup plus sensible à ces derniers et au grand nombre de variables explicatives (du notamment à la discrétisation des variables).  
Le Ridge n'est cependant pas suffisant pour expliquer complètement la demande de vélos partagés.

```python
class CustomRidge(BaseEstimator):
    """
    CustomRidge utilise le Ridge pour predire separement
    la valeur du logarithme de variables 'registered' et 'casual'.
    Les valeurs sont ensuite repassees a l'exponentiel et la variable 
    'count' correspond à la somme de 'registered' et 'casual'
    """
    
    def __init__(self):
        self.casual_lr = Ridge()
        self.registered_lr = Ridge()

    def fit(self, X, y):
        """
        entrainement de deux modeles Ridge separes
        sur le logarithme des variables 'casual' et 'registered'
        """
        y_r_log = np.log(y[:, 0] + 1)
        y_c_log = np.log(y[:, 1] + 1)
        self.registered_lr.fit(X, y_r_log)
        self.casual_lr.fit(X, y_c_log)

    def predict(self, X):
        """
        prediction a partir de deux modeles Ridge separes
        du logarithme des variables 'casual' et 'registered'
        La prevision de la variable 'count' correspond à la somme
        de l'exponentiel des predictions pour les usagers abonnes
        et celui des usagers occasionnels
        """
        lr_pred = np.zeros((X.shape[0], 3))
        lr_pred[:, 0] = np.exp(self.registered_lr.predict(X)) - 1
        lr_pred[:, 1] = np.exp(self.casual_lr.predict(X)) - 1
        lr_pred[:, 2] = lr_pred[:, 0] + lr_pred[:, 1]

        # on predit 0 pour lorsque le Ridge predit une valeur negative
        lr_pred[lr_pred < 0] = 0
        
        return lr_pred

    def score(self, X, y):
        """
        RMSLE calculee uniquement sur la variable 'count'
        """
        diff = np.log(self.predict(X)[:, 2] + 1) - np.log(y[:, 2] + 1)
        mean_error = np.square(diff).mean()

        return np.sqrt(mean_error)
```

#### Gradient Boosting

```python
class CustomGradientBoosting(BaseEstimator):
    """
    CustomGradientBoosting utilise le Gradient Boosting pour predire separement
    la valeur du logarithme de variables 'registered' et 'casual'.
    Les valeurs sont ensuite repassees a l'exponentiel et la variable 
    'count' correspond à la somme de 'registered' et 'casual'
    """
    
    def __init__(self):
        self.casual_gb = GradientBoostingRegressor(n_estimators=117,
                                                   learning_rate=0.05,
                                                   max_depth=8,
                                                   min_samples_leaf=10)
        self.registered_gb = GradientBoostingRegressor(n_estimators=117,
                                                       learning_rate=0.055,
                                                       max_depth=9,
                                                       min_samples_leaf=10)

    def fit(self, X, y):
        """
        entrainement de deux modeles de Gradient Boosting separes
        sur le logarithme des variables 'casual' et 'registered'
        """
        y_r_log = np.log(y[:, 0] + 1)
        y_c_log = np.log(y[:, 1] + 1)
        self.registered_gb.fit(X, y_r_log)
        self.casual_gb.fit(X, y_c_log)
        self.feature_importances_ = \
            np.vstack([self.registered_gb.feature_importances_,
                       self.casual_gb.feature_importances_])

    def predict(self, X):
        """
        prediction a partir de deux modeles Gradient Boosting separes
        du logarithme des variables 'casual' et 'registered'
        La prevision de la variable 'count' correspond à la somme
        de l'exponentiel des predictions pour les usagers abonnes
        et celui des usagers occasionnels
        """
        gb_pred = np.zeros((X.shape[0], 3))
        gb_pred[:, 0] = np.exp(self.registered_gb.predict(X)) - 1
        gb_pred[:, 1] = np.exp(self.casual_gb.predict(X)) - 1
        gb_pred[:, 2] = gb_pred[:, 0] + gb_pred[:, 1]
        gb_pred[:, 2][gb_pred[:, 2] < 0] = 0
        
        return gb_pred

    def score(self, X, y):
        """
        RMSLE calculee uniquement sur la variable 'count'
        """
        diff = np.log(self.predict(X)[:, 2] + 1) - np.log(y[:, 2] + 1)
        mean_error = np.square(diff).mean()

        return np.sqrt(mean_error)
```

Le Gradient Boosting est une methode d'ensemble learning qui a fait ses preuves dans de nombreuses competitions. Ils combine de nombreux (un peu plus d'une centaine dans notre cas) prédicteurs faibles (comme des arbres de décision) afin de tenir compte de la complexité du phénomène. Sa particularité tient dans le fait que les prédicteurs faibles se concentrent sur les données pour lesquelles les erreurs des predicteurs précedents sont les plus importantes. L'optimisation se fait via une descente de gradient. L'algorithme permet d'obtenir des résultats beaucoup plus satisfaisant que les deux algorithmes précédents puisque la CV RMSLE est proche de $0.37~ (+/-0.10)$. Elle atteint même $0.3493$ sur le test effectué sur les 6 derniers mois du dataset. La CV RMSE est alors de $57~ (+/-17)$.  
  
Le Gradient Boosting est le modèle qui m'a permis d'obtenir les meilleures prédictions pour la demande de vélos partagés. C'est donc ce modèle de prédiction que je retiendrais. Cependant, ce modèle est beaucoup plus couteux à mettre en oeuvre que le Ridge. Il suppose plus de puissance et de temps de calcul et il est plus compliqué à interprêter (bien qu'il permettent de calculer l'importance relative de chacune des variables).

### Le critère de performance utilisé.

Un critère de performance adapté à la prédiction du nombre des locations de vélos est la minimisation de la racine du carré moyen des écarts entre le logarithme de la prédiction et celui du résultat observé dans la réalité. Cette mesure est notée **RMSLE (Root Mean Squared Logarithmic Error)** dans ce notebook et est définie de la manière suivante :  
$$RMSLE(\vec{a}, \vec{p}) = \sqrt{\frac{1}{n}\sum_{i=1}^n(\log(p_t + 1) - \log(a_t + 1))^2}$$
avec :  
- n : le nombre de prédictions réalisées
- $p_t$ : la prédiction du nombre de locations à l'instant $t$ (vecteur $\vec{p}$)
- $a_t$ : le nombre réel de locations à l'instant $t$ (vecteur $\vec{a}$)
- $\log(x)$ : le logarithm néperien
  
Ce critère est basé sur la proportion de locations prédites par rapport à celles qui sont constatées. En effet, $$\log(p_t + 1) - \log(a_t + 1) = \log{\frac{p_t + 1}{a_t + 1}}$$.  

Le logarithme permet ici d'évaluer l'ordre de grandeur des prédictions et des observations afin de mesurer leur adéquation. Le carré permet de renvoyer une mesure positive de l'erreur que le nombre de locations soit sur-estimé ou sous-estimé. Le RMSLE correspond alors à la racine du carré moyen des logarithmes des ratios $\frac{p_t + 1}{a_t + 1}$. Sur le graphique ci-dessous, on peut vérifier que si ce ratio vaut 1 ($p_t = a_t$), alors l'erreur est nulle. Si on sous-estime le nombre de locations en moyenne de motié ($p_t = \frac{a_t}{2}$ et le ratio vaut $0.5$), la RMSLE vaut environ $0.67$. La mesure est alors similaire si les prédictions sont deux fois plus élevées que le nombre de locations observé (soit $p_t = 2 a_t$ et le ratio vaut $2$). Le RMSLE renvoit de manière générale une erreur similaire pour $p_t = \frac{a_t}{\alpha}$ et $p_t = \alpha a_t$ (où $\alpha$ correspondrait au ratio).

J'ai également indiqué pour chacun des modèles la racine de la moyenne des erreurs au carré (**Root-Mean-Square-Error, RMSE**) qui est plus fréquemment utilisée. Elle peut être interprêtée comme la distance euclidienne entre la prédiction du nombre de vélos loués par heure et la réalité. On ne peut pas utiliser directement l'écart moyen entre la prédiction et la quantité constatée car sous-estimer le nombre de vélos qui seront loués à un certain moment compenserait une sur-estimation de ce nombre à un autre moment.  
  
Enfin, ne disposant pas d'information sur le rapport entre le coût d'une sur-estimation de vélos nécessaires (qui suppose l'achat et l'entretien de vélos qui seront rarement utilisés) et celui d'une sous-estimation des locations (qui risque de peser sur le nombre d'abonnés au service), je considère ici que les deux erreurs sont équivalentes (en proportion et pas en nombre de vélos). Il pourrait être judicieux d'adapter le critère de performance afin de tenir compte de la structure de coût lors de l'oprimisation. On pourrait par exemple donner plus de poids aux sous-estimation du nombre de locations dans l'erreur afin que le modèle réalise des prédictions plus prudentes dans ce sens.

```python
# pred_ratio correspond au ratio predictions / nombre de locations observees - 1
pred_ratio = np.linspace(-1.0, 1.0, num=100)
rmsle_score = np.array(list(map(lambda pr: rmsle((1 - pr) * y_test[:, 2], y_test[:, 2]), pred_ratio)))
rmse_score = np.array(list(map(lambda pr: rmse((1 - pr) * y_test[:, 2], y_test[:, 2]), pred_ratio)))

fmt = '%.0f%%' # Format de l'axe des abscisses
xticks = mtick.FormatStrFormatter(fmt)

# graphique RMSLE
plt.figure(figsize=(12, 4))
ax1 = plt.subplot(1, 2, 1)
plt.plot(pred_ratio * 100, rmsle_score, label='RMSLE')
ax1.xaxis.set_major_formatter(xticks)
plt.xlabel("$p_t/a_t-1$ (en %)")
plt.ylabel('RMSLE')
plt.title("RMSLE en fonction du pourcentage d'erreur de prédiction")

# graphique RMSLE
ax2 = plt.subplot(1, 2, 2)
plt.plot(pred_ratio * 100, rmse_score, label='RMSE')
plt.xlabel("$p_t/a_t-1$ (en %)")
ax2.xaxis.set_major_formatter(xticks)
plt.ylabel('RMSE')
plt.title("RMSE en fonction du pourcentage d'erreur de prédiction")
plt.show()
```
